```python
import lightkurve as lk
%matplotlib inline
```

Cel badania

```python
TIC = "TIC 284475976"
```

```python
data_1 = lk.search_lightcurve(TIC)
data_1
```

```python
data_2 = lk.search_lightcurve(TIC, author="SPOC")
data_2
```

```python
data_3 = lk.search_lightcurve(TIC, author="SPOC", sector=23)
data_3
```

```python
lc = data_3.download()
lc.plot(linewidth = 0, marker = ".", color = "midnightblue", alpha = 0.3)
```

```python
# normalizowanie pojedynczego sektora
lc.normalize().plot()
```

```python
TIC_2 = "TIC 55525572"
data_2 = lk.search_lightcurve(TIC_2, author="SPOC")
data_2
```

```python
new_data = data_2[:4]
new_data
```

```python
lc_new_data = new_data.download_all()
lc_new_data.plot()
```

```python
# normalizacja danych z wielu sektorów przy użyciu metody stitch()
lc_new_data_normalized = lc_new_data.stitch()
lc_new_data_normalized.plot()
```
